```python
import geopandas as gpd
from pyproj import Proj
```

```python
water_points = gpd.read_file("web_chennai_water/points.geojson")
water_points["id"]= ""
```

```python
for i in range(0,len(water_points)): water_points["id"][i] = i
```

```python
water_points.to_file("web_chennai_water/points.geojson",driver="GeoJSON")
```

```python
water_polygons = gpd.read_file("web_chennai_water/polygons.geojson")
```

```python
water_points["geometry1"] = 0
```

```python
len(water_points)
```

```python
for index in range(0,len(water_points)):
    points = water_polygons.intersects(water_points.geometry[index])
    points_intersected = [i for i, e in enumerate(points) if e == True]
    water_points["geometry1"][index] = water_polygons.geometry[points_intersected[0]]
    x,y = water_points.geometry[index].coords.xy
    water_points["geometry"][index] = str([x[0],y[0]])
```

```python
water_points.columns=water_points.columns.str.replace('geometry','points')
water_points.columns=water_points.columns.str.replace('points1','geometry')
```

```python
water_points= water_points.to_crs({'init': 'epsg:3857'})
```

```python
water_points["area_in_m"] = water_points.geometry.area
```

```python
water_points= water_points.to_crs({'init': 'epsg:4326'})
```
